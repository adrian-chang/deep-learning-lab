# DataLoader

* epoch = outer loop
* batch size = amount of data to feed in
* iterations = 1 iteration per batch

```python
import torch as t
from torch.utils.data import Dataset, DataLoader
import numpy as np
```

```python
class DiabetesDataset(Dataset):
    def __init__(self):
        all_data = all_data = np.loadtxt('./data/diabetes.csv', delimiter=',', dtype=np.float32)
        self.len = all_data.shape[0]
        self.x_data = t.from_numpy(all_data[:, 0:-1])
        self.y_data = t.from_numpy(all_data[:, -1])
    def __getitem__(self, index):
        return self.x_data[index], self.y_data[index]
    def __len__(self):
        return self.len
    
dataset = DiabetesDataset()
train_loader = DataLoader(dataset = dataset, batch_size=32, shuffle=True, num_workers=2)
```

```python
class ModelDiabetes(t.nn.Module):
    def __init__(self):
        super(ModelDiabetes, self).__init__()
        self.l1 = t.nn.Linear(8, 6)
        self.l2 = t.nn.Linear(6, 4)
        self.l3 = t.nn.Linear(4, 1)
        self.sigmoid = t.nn.Sigmoid()
    def forward(self, x):
        out_1 = self.sigmoid(self.l1(x))
        out_2 = self.sigmoid(self.l2(out_1))
        return self.sigmoid(self.l3(out_2))
```

```python
model_diabetes = ModelDiabetes()
cri_diabetes = t.nn.BCELoss(size_average=True)
opt_diabetes = t.optim.SGD(model_diabetes.parameters(), lr=0.1)
```

```python
for times in range(10):
    for i, batch in enumerate(train_loader):
        features, labels = batch
        y_pred = model_diabetes.forward(features)
        loss = cri_diabetes(y_pred, labels) # labels causes warning, it's 23 x 0, when y_pred is 23 x 1
        print(f'epoch = {i}, loss = {loss.item()}') # this deviates from the lecture
        opt_diabetes.zero_grad()
        loss.backward()
        opt_diabetes.step()
```
